## Calculate means for visualization

```python
story_id_list = result_df["story_id"].unique()
mean_df = pd.concat([result_df[result_df["story_id"] == story_id].mean() for story_id in story_id_list], axis=1).T
mean_df["story_id"] = story_id_list

id_with_type = result_df[["story_id","Title","type"]].drop_duplicates()
mean_df = pd.merge(mean_df, id_with_type, on='story_id')
```

## Visualize "Emotional Flow"

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import sys
sys.path.append("../src/")
from emotionplot import plot_result, print_info

N = len(mean_df)

for n in range(N):
    sample_result = mean_df.iloc[[n]]
    
    print("Story ID:", sample_result.loc[:, "story_id"].values[0])
    print("Type:", sample_result.loc[:, "type"].values[0])
    print("Rating:", sample_result.loc[:, "rating"].values[0])
    
    fig = plot_result(sample_result)
    plt.show()
```
